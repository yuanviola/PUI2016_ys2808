```python
import numpy as np
import pandas as pd
import pylab as pl
import seaborn as sns
%pylab inline
```

```python
data=pd.read_csv('Water_Quality_complaints.csv')
data.head()
```

```python
data.rename(columns={'Incident Zip':'zipcode'},inplace=True)
```

```python
data.zipcode.dropna(inplace=True)
```

```python
data.head()
```

```python
a=data['zipcode'].groupby(data['zipcode']).count()
a
```

```python
pl.figure(figsize=(50,5))
a.plot(kind='bar')
pl.xlabel('Incident Zipcode',fontsize=20)
pl.xticks(fontsize=10)
pl.ylabel('Complaints counted according to zipcode')
pl.title('NYC water complaints counts by zipcode')
#please double click  the plot, it will be bigger and clear!!
```
