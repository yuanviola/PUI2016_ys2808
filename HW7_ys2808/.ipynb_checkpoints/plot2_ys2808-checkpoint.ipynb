```python
import numpy as np
import pandas as pd
import pylab as pl
import seaborn as sns
%pylab inline
```

```python
data=pd.read_csv('Water_Quality_complaints.csv')
data.head()
```

```python
data.rename(columns={'Incident Zip':'zipcode'},inplace=True)
```

```python
data.zipcode.dropna(inplace=True)
```

```python
data.head()
```

```python
a=data['zipcode'].groupby(data['zipcode']).count()
a
```

```python
a.sort?
```

```python
a.sort(ascending=False)
a=pd.DataFrame(a)
a.rename(columns={'zipcode':'count'},inplace=True)
a.iloc[:10]
```

```python
pl.legend?
```

```python
pl.figure(figsize=(12,8))
a.iloc[0:10].plot(kind='bar')
pl.xlabel('Incident Zipcode',fontsize=20)
pl.xticks(fontsize=10)
pl.ylabel('Complaints counts',fontsize=20)
pl.title('NYC water complaints counts by zipcode',fontsize=20)
```
